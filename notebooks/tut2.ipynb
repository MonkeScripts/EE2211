```python
df = pd.read_csv("../data/AnnualMotorVehiclePopulationbyVehicleType.csv")
print('--- head ---')
print(df.head(50))
```

```python
year = df['year']
category = df['category']
vehtype = df['type']
number = df["number"]

df_private_buses = df.loc[df['type']=="Private buses"]


plt.figure(figsize=(9, 4))
plt.plot(df_private_buses['year'], df_private_buses['number'], label='Private buses',   marker='^')
plt.xlabel('Year')
plt.ylabel('Number of vehicles')
plt.title('Number of vehicles over the years')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from pandas.plotting import scatter_matrix

# 1. Load data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 2. Create the scatter matrix
# 'c' colors the points by species (target)
# 'diagonal' can be 'hist' (default) or 'kde' (kernel density estimation)
scatter_matrix(df, c=iris.target, figsize=(10, 10), marker='o', 
               s=40, alpha=0.8, diagonal='hist')

plt.show()
```
