```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poly_features(X, d):
    """Build polynomial feature matrix [x, x^2, ..., x^d] with bias."""
    cols = [X ** k for k in range(1, d + 1)]
    X_poly = np.hstack(cols)
    N_rows = X_poly.shape[0]
    return np.hstack((np.ones((N_rows, 1)), X_poly))

plt.figure(figsize=(12, 4))

for i, d in enumerate([3]):
    X_b = poly_features(X, d)
    print(f'X_b shape: {X_b.shape}')
    w = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
    mse = np.mean((X_b @ w - y) ** 2)

    X_plot_b = poly_features(x_plot, d)
    y_plot   = X_plot_b @ w

    ax = plt.subplot(1, 3, i + 1)
    ax.scatter(X, y, alpha=0.5, s=20, label='data')
    ax.plot(x_plot, y_plot, color='red', linewidth=2, label=f'deg {d}')
    ax.set_title(f'Degree {d}  (MSE={mse:.2f})')
    ax.legend()
    print(f'Degree {d}: coefficients = {w.ravel().round(3)},  MSE = {mse:.4f}')

plt.tight_layout()
plt.show()
```
