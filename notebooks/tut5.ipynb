```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.asarray([-10, -8, -3, -1, 2, 8]).reshape(-1, 1)
y = np.asarray([5, 5, 4, 3, 2, 2])
N = X.shape[0]
X.shape
```

```python
X_b = np.hstack((np.ones((N, 1)), X))              # shape (N, 2)
# Solve: w = [bias, slope]
w_b = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y     # shape (2, 1)

y_pred_b = X_b @ w_b
mse_b = np.mean((y_pred_b - y) ** 2)

print(f'Intercept w[0] = {w_b[0]:.4f}  (true ≈ 5.0)')
print(f'Slope     w[1] = {w_b[1]:.4f}  (true ≈ 3.0)')
print(f'Train MSE      = {mse_b:.4f}')

plt.figure(figsize=(7, 4))
plt.scatter(X, y, alpha=0.6, label='data')
plt.plot(X, y_pred_b, color='red', linewidth=2,
         label=f'fit: y={w_b[1]:.2f}x + {w_b[0]:.2f}')
plt.legend()
plt.title('Linear Regression (with bias)')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

w = np.linalg.inv(X.T @ X) @ X.T @ y              # shape (1, 1)

y_pred = X @ w
mse = np.mean((y_pred - y) ** 2)
print(f'Slope w   = {w[0]:.4f}  (true ≈ 2.0)')
print(f'Train MSE = {mse:.4f}')

plt.figure(figsize=(7, 4))
plt.scatter(X, y, alpha=0.6, label='data')
plt.plot(X, y_pred, color='red', linewidth=2, label=f'fit: y={w[0]:.2f}x')
plt.legend()
plt.title('Linear Regression (no bias)')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```
